```python
events = pd.read_csv("data/events.csv", index_col=0)

events["last_clickout_of_session"] = 0

events.loc[
    events[events["action_type"] == "clickout item"]
    .sort_values(["session_id", "step"], ascending=False)
    .drop_duplicates("session_id")
    .index,
    "last_clickout_of_session"
] = 1

events.to_csv("data/events.csv")
```
